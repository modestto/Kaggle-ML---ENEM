```python
#Separando as variáveis dependentes e independentes do nosso modelo
y1=train_clean.loc[:,'NU_NOTA_CN']
y2=train_clean.loc[:,'NU_NOTA_CH']
y3=train_clean.loc[:,'NU_NOTA_LC']
y4=train_clean.loc[:,'NU_NOTA_MT']
y5=train_clean.loc[:,'NU_NOTA_REDACAO']
train_clean = train_clean.drop(columns=['NU_NOTA_CN','NU_NOTA_CH','NU_NOTA_LC','NU_NOTA_MT','NU_NOTA_REDACAO'])
```

```python
scaler=StandardScaler()
scaler.fit(train_clean)
train_scaled=scaler.transform(train_clean)
test_scaled=scaler.transform(test_clean)
```

```python
X_train_scaled = pd.DataFrame(train_scaled, columns=train_clean.columns)
```
