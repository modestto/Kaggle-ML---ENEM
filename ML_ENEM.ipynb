# Modelo

```python
#Separando as variáveis dependentes e independentes do nosso modelo

x1 = train_clean.loc[:,['NU_NOTA_CH','NU_NOTA_LC','NU_NOTA_MT','NU_NOTA_REDACAO']]
y1=train_clean.loc[:,'NU_NOTA_CN']

x2 = train_clean.loc[:,['NU_NOTA_CN','NU_NOTA_LC','NU_NOTA_MT','NU_NOTA_REDACAO']]
y2=train_clean.loc[:,'NU_NOTA_CH']

x3 = train_clean.loc[:,['NU_NOTA_CH','NU_NOTA_CN','NU_NOTA_MT','NU_NOTA_REDACAO']]
y3=train_clean.loc[:,'NU_NOTA_LC']

x4 = train_clean.loc[:,['NU_NOTA_CH','NU_NOTA_LC','NU_NOTA_CN','NU_NOTA_REDACAO']]
y4=train_clean.loc[:,'NU_NOTA_MT']

x5 = train_clean.loc[:,['NU_NOTA_CH','NU_NOTA_LC','NU_NOTA_MT','NU_NOTA_CN']]
y5=train_clean.loc[:,'NU_NOTA_REDACAO']

```

```python
x_train = sm.add_constant(x1) 
model1 = sm.OLS(y1.astype(float), x_train.astype(float)).fit()

x_train = sm.add_constant(x2)
model2 = sm.OLS(y2.astype(float), x_train.astype(float)).fit()

x_train = sm.add_constant(x3)
model3 = sm.OLS(y3.astype(float), x_train.astype(float)).fit()

x_train = sm.add_constant(x4)
model4 = sm.OLS(y4.astype(float), x_train.astype(float)).fit()

x_train = sm.add_constant(x5)
model5 = sm.OLS(y5.astype(float), x_train.astype(float)).fit()
```
